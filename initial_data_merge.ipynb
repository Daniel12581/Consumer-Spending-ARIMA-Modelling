```python
import pandas as pd
```

```python
df_co2 = pd.read_csv("data/owid-co2-data.csv")
df_ce = pd.read_csv("data/renewable-energy-gen.csv")
```

```python
df_co2.head()
```

```python
df_ce.head()
```

```python
df_co2 = df_co2[df_co2["country"] == "United States"]
df_ce.rename(columns = {"Entity": "country", "Year": "year"}, inplace = True)
df_ce = df_ce[df_ce["country"] == "United States"]
```

```python
df_co2.head()
```

```python
df_ce.head()
```

```python
df = df_co2.merge(df_ce, on = ["country", "year"])
```

```python
df.head()
```

```python
df.rename(columns = {"Solar generation - TWh": "solar_generation", "Wind generation - TWh": "wind_generation", "Hydro generation - TWh": "hydro_generation", "Other renewables (including geothermal and biomass) electricity generation - TWh": "other_generation"}, inplace = True)
```

```python
df = df[["year", "population", "gdp", "co2", "solar_generation", "wind_generation", "hydro_generation", "other_generation"]]
```

```python
df.to_csv("data/merged_data.csv", index = False)
```
